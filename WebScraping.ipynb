```python
before_clean = pd.read_excel('lazada_โต๊ะปรับระดับ.xlsx')
before_clean
```

```python
before_clean_price = before_clean['Price']
before_clean_price
```

```python
cleaned_prices = []  # Initialize an empty list to store the cleaned prices

for price in before_clean_price:
    # Remove the "฿" symbol and any commas (for thousands separator)
    cleaned_price = price.replace("฿", "").replace(",", "")
    
    # Append the cleaned price as a float to the list
    cleaned_prices.append(float(cleaned_price))

# Now, cleaned_prices contains a list of floats
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Product Name': ['A', 'B'],
    'Price': [10, 99],
    'Amount(pcs)': [50, 100]
}

# Create a DataFrame from the data
df = pd.DataFrame(data)

# Bar chart
plt.figure(figsize=(8, 6))
plt.bar(df['Product Name'], df['Price'], label='Price', alpha=0.7)
plt.bar(df['Product Name'], df['Amount(pcs)'], label='Amount(pcs)', alpha=0.7)
plt.xlabel('Product Name')
plt.ylabel('Value')
plt.title('Product Selling Data')
plt.legend()
plt.show()

# Line chart (assuming time information)
# If you have a time column, replace 'Time' with the actual column name.
# Example:
# df['Time'] = ['Jan', 'Feb']
# Then, use df['Time'] as the x-axis.
plt.figure(figsize=(8, 6))
plt.plot(df['Product Name'], df['Price'], label='Price', marker='o')
plt.plot(df['Product Name'], df['Amount(pcs)'], label='Amount(pcs)', marker='o')
plt.xlabel('Product Name')
plt.ylabel('Value')
plt.title('Product Selling Data Over Time')
plt.legend()
plt.grid(True)
plt.show()

# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['Price'], df['Amount(pcs)'], c='blue', marker='o')
plt.xlabel('Price')
plt.ylabel('Amount(pcs)')
plt.title('Price vs. Amount(pcs)')
plt.grid(True)
plt.show()

# Pie chart
plt.figure(figsize=(8, 6))
plt.pie(df['Price'], labels=df['Product Name'], autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Price by Product')
plt.show()

# Heatmap (assuming you have more data)
# Create a pivot table for the heatmap
heatmap_data = df.pivot_table(index='Product Name', columns='//comment: Head Column', values='Price')
plt.figure(figsize=(8, 6))
plt.imshow(heatmap_data, cmap='YlGnBu', interpolation='nearest')
plt.colorbar()
plt.xticks(range(len(heatmap_data.columns)), heatmap_data.columns, rotation=45)
plt.yticks(range(len(heatmap_data.index)), heatmap_data.index)
plt.title('Heatmap of Price by Product')
plt.show()
```
